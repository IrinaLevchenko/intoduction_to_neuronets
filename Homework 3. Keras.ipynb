```python
import numpy as np
import pandas as pd
import mnist
import keras

from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.utils import to_categorical
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Import images
train_images = mnist.train_images()
train_labels= mnist.train_labels()

test_images = mnist.test_images()
test_labels = mnist.test_labels()

print(train_images.shape)
print(test_images.shape)
```

```python
# Normalize the images
train_images = (train_images / 255) - 0.5
test_images = (test_images / 255) - 0.5

# Flatten the images
train_images = train_images.reshape((-1, 784))
test_images = test_images.reshape((-1, 784))

print(train_images.shape)
print(test_images.shape) 
```

```python
def create_model(optimizer='adam'):
    # Build the model (architecture)
    model = Sequential([
        Dense(64, activation='relu', input_shape=(784,)),
        Dense(32, activation='relu'),
        Dense(10, activation='softmax')
    ])
    # Compile the model (задаём параметры)
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

## 1ое действие:

Подбираю лучшие параметры для модели с помощью гридсёрч

```python
batch_size = [50, 500, 1000]
epochs = [5, 10, 15]
optimizer = ['SGD', 'RMSprop', 'Adam']

model = KerasClassifier(create_model)
params= dict(batch_size=batch_size, epochs=epochs, optimizer=optimizer)
grid = GridSearchCV(model, params, refit=False, cv=3, scoring='accuracy')

grid_result = grid.fit(train_images, to_categorical(train_labels))
print(grid_result.best_score_, grid_result.best_params_)
```

```python
print(grid_result.best_score_, grid_result.best_params_)
```

Создаю и обучаю модель на параметрах, подобранных гридсёрч

```python
model_1 = create_model(optimizer='RMSprop')

# Train the model
model_1.fit(train_images, to_categorical(train_labels), epochs=15, batch_size=50)

# Evaluate the model
loss, metrics = model_1.evaluate(test_images, to_categorical(test_labels))
```

```python
print(f'model_1: Loss - {np.round(loss,2)}, Metrics - {np.round(metrics,3)}')
```

Получила хорошие метрики

Фиксирую и делаю предсказания, вывожу результат, чтобы убедиться в том как хорошо работает модель
    

```python
# Predict on the first 10 test images.
predictions = model_1.predict(test_images[:20])

# Print our model's predictions.
print(f'Predictions: {np.argmax(predictions, axis=1)}')

# Check our predictions against the ground truths.
print(f'True labels: {test_labels[:20]}')
```

## 2ое действие:

Изменю архитектуру модели

```python
def create_model(optimizer='RMSprop'):
    # Build the model (architecture)
    model = Sequential([
        Dense(128, activation='relu', input_shape=(784,)),
        Dense(64, activation='relu'),
        Dense(10, activation='softmax')
    ])
    # Compile the model (задаём параметры)
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
model_2 = create_model()

# Train the model
model_2.fit(train_images, to_categorical(train_labels), epochs=15, batch_size=50)

# Evaluate the model
loss, metrics = model_2.evaluate(test_images, to_categorical(test_labels))
```

```python
print(f'model_2: Loss - {np.round(loss,2)}, Metrics - {np.round(metrics,3)}')
```

Вижу, что метрики немного улучшились. Хотя не значительно
